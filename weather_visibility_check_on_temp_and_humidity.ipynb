```python
path = "us_weather/sitka_weather_07-2014.csv"

import csv

with open(path) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    print(header_row)
```

```python
import pandas as pd

df_w = pd.read_csv(path)
df_w[:5]
```

```python
import seaborn as sb

mean_temp = df_w['Mean TemperatureF']
mean_temp.plot(kind = 'line')
```

```python
df_w[['Mean TemperatureF',' Mean VisibilityMiles']].plot(kind='barh')
```

```python
# Center the data in mean temperature and mean visibility to see pattern 

df_mtv = df_w[['Mean TemperatureF', ' Mean VisibilityMiles']]
df_center = df_mtv.apply(lambda x: x - x.mean())
df_center.plot(title='Plot of Temperature Flunctuations with Visibility')
```

```python
"""Temperature flunctuation is not essentially followed on by Visibility, there seems to be other factor influencing visibility.
Check for humidity and atmospheric pressure """

df_mtvh = df_w[['Mean TemperatureF', ' Mean VisibilityMiles' ,' Mean Humidity']]
df_cmtvh = df_mtvh.apply(lambda x: x - x.mean())

thv_plot = df_cmtvh.plot(title='Plot of Temperature Flunctuations with Humidity and Visibility')
thv_plot.hlines(y=0, xmin=0, xmax=31, color='black')
```

```python
import matplotlib.pyplot as plt

df_temp = df_w[['AKDT','Max TemperatureF', 'Min TemperatureF']]
df_temp.columns = ('date', 'high', 'low')
#fig, axes = plt.subplots(1,2, sharey=True)
#sb.lineplot(x = 'date', y='high', data=df_temp)
df_temp.plot(kind='line', x='date', rot=45)
```

```python
path2 = 'us_weather/sitka_weather_2014.csv'

df_w2014 = pd.read_csv(path2)
df_w2014[:5]
```

```python
df_2014_thl = df_w2014[['AKST','Max TemperatureF', 'Min TemperatureF']]
df_2014_thl.columns = ('date', 'high', 'low')
df_2014_thl.plot(x='date', kind='line', rot=45, title="Sitka Weather 2014: Daily High and Low Temperature in 2014")
```

```python
# Plot of Temperature flunctuation with Humidity

df_w2014_th = df_w2014[['AKST', 'Mean TemperatureF', ' Mean Humidity', ' Mean VisibilityMiles']]
df_w2014_th.columns = ('date','m_Temperature','m_Humidity','m_Visibility')
df_centered = df_w2014_th.iloc[:,1:4].apply(lambda x: x - x.mean())
df_centered['date'] = pd.to_datetime(df_w2014_th['date'])
plot = df_centered.plot(x='date', rot=45, title="Sitka Weather 2014: Temperature Flunctuation with Humidty")
plot.hlines(y=0, color='black', xmin='2014-01-01', xmax='2014-12-31')
```

```python
""" Doing analysis on Death Valley Weather Data """
path3 = 'us_weather/death_valley_2014.csv'

df_dv2014 = pd.read_csv(path3)
df_dv2014[:3]
```

```python
pd.to_datetime(df_dv2014['PST'])
df_hlt = df_dv2014[['Max TemperatureF', 'Min TemperatureF']]
df_hlt.loc[:,'date'] = pd.to_datetime(df_dv2014['PST'])
df_hlt.set_index('date', inplace=True)
df_hlt
```

```python
""" Grouping High and Low Temperature measurements by week""" 
wkly_df_hlt = df_hlt.resample('W').asfreq()
wkly_df_hlt.plot(rot=45)
```
