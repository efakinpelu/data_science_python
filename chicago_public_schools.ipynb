```python
#df_admit['Program_Type'] # program type in chicago schools
import numpy as np
# write the below for loop as pan.Series function
#pan.Series(x.replace(df_admit.Short_Name, '') for x in df_admit.Program_Type)

for ind in df_admit.index:
    r = df_admit['Short_Name'][ind]
    s = df_admit['Program_Type'][ind]
    t = s.replace(r, '').strip(' - ')
    df_admit.loc[:,('Program_Code')][ind] = t.lower() # avoid SettingWithCopyWarning by using .loc method of the dataframe
    
df_admit
```

```python
# retrieve the High Schools in STEM
#stem_school = df_admit[(df_admit.Program_Group == 'STEM') & (df_admit.Primary_Category == 'ES')]
stem_school = df_admit[(df_admit.Program_Group == 'STEM')]
grp_stem = stem_school.groupby(['Program_Code', 'Primary_Category'])
grp_stem.size() # view the Category of STEM Education Offered in Chicago
```
